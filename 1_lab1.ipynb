```python
from dotenv import load_dotenv
from openai import AsyncOpenAI
from agents import Agent, Runner, trace, OpenAIChatCompletionsModel
import os
```

```python
# The usual starting point

load_dotenv(override=True)
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
google_api_key = os.getenv("GEMINI_API_KEY")
if google_api_key:
    print('Google api key exists!')
```

```python
gemini_client = AsyncOpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)

gemini_model = OpenAIChatCompletionsModel(model = 'gemini-2.5-flash', openai_client=gemini_client)
```

```python

# Make an agent with name, instructions, model

agent = Agent(name="Jokester", instructions="You are a joke teller", model = gemini_model)
```

```python


os.environ["OPENAI_API_KEY"] = os.getenv("GEMINI_API_KEY")
os.environ["OPENAI_BASE_URL"] = "https://generativelanguage.googleapis.com/v1beta/openai/"

agent = Agent(
    name="Jokester",
    instructions="You are a joke teller",
    model="gemini-2.5-flash"
)

print("Gemini key loaded:", os.getenv("OPENAI_API_KEY")[:5])
```

```python
result = await Runner.run(agent, "Tell me a joke about AI Autonomouse agents")
print(result.final_output)
```

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

with trace("Telling a joke"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
    print(result.final_output)
```
